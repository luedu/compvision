# Image inpainting
https://www.pyimagesearch.com/2020/05/18/image-inpainting-with-opencv-and-python/

```python
import cv2
import os
import matplotlib.pyplot as plt
```

## Configuration

```python
image_folder = 'examples/'
method = "ns"  # Inpainting algorithm to use: "telea", "ns"
radius = 3  # Inpainting radius
```

## Initialization

```python
# Initialize the inpainting algorithm to be the Telea et al. method
flags = cv2.INPAINT_TELEA

# Check to see if we should be using the Navier-Stokes (i.e., Bertalmio et al.) method for inpainting
if method == "ns":
    flags = cv2.INPAINT_NS
    
# Load the (1) input image (i.e., the image we're going to perform
# inpainting on) and (2) the  mask which should have the same input
# dimensions as the input image -- zero pixels correspond to areas
# that *will not* be inpainted while non-zero pixels correspond to
# "damaged" areas that inpainting will try to correct
image = []
mask = []
for image_number in range(1, 4):
    image.append(cv2.imread(os.path.join(image_folder, f"example0{image_number}.png")))
    mask.append(cv2.cvtColor(cv2.imread(os.path.join(image_folder, 
                                                     f"mask0{image_number}.png")),cv2.COLOR_BGR2GRAY))
```

## Inpainting

```python
# Perform inpainting using OpenCV
output = []
for i in range(3):
    output.append(cv2.inpaint(image[i], mask[i], radius, flags=flags))


    plt.figure(figsize=(15, 15))
    plt.subplot(131)
    plt.imshow(cv2.cvtColor(image[i], cv2.COLOR_BGR2RGB))
    plt.title('Original image')
    plt.subplot(132)
    plt.imshow(mask[i])
    plt.title('Inpainting mask')
    plt.subplot(133)
    plt.imshow(cv2.cvtColor(output[i], cv2.COLOR_BGR2RGB))
    plt.title('Inpainting result')
```
